```python
import cv2
import math

# Initialize the YOLO model
model = YOLO('models/fall_model.pt')

# Load class names from the file
classnames = []
with open('classes/class.txt', 'r') as f:
    classnames = f.read().splitlines()

# Open the webcam
cap = cv2.VideoCapture(0)

# Initialize dictionaries to store bounding box coordinates for each class
bounding_boxes = {
    'Fall Detected': [],
    'NoFall Detected': []  # Add other class names as needed
}

while True:
    # Read a frame from the webcam
    ret, frame = cap.read()
    
    # Resize the frame
    frame = cv2.resize(frame, (980, 740))
    
    # Process the frame through the YOLO model
    results = model(frame)
    
    # Initialize variables to track the highest confidence and its corresponding class
    highest_confidence = -1
    highest_confidence_class = None
    
    
    # Iterate over the detection results
    for info in results:
        parameters = info.boxes
        for box in parameters:
            print(box)
            x1, y1, x2, y2 = box.xyxy[0]
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
            confidence = box.conf[0]
            class_detect = box.cls[0]
            class_detect = int(class_detect)
            class_detect = classnames[class_detect]
            conf = math.ceil(confidence * 100)
            
            # Check if the current detection has a higher confidence than the previously recorded highest confidence
            if conf > highest_confidence:
                highest_confidence = conf
                highest_confidence_class = class_detect
            
            # Store the bounding box coordinates for the detected class
            bounding_boxes[class_detect].append([x1, y1, x2, y2])
            
            # Draw rectangle and label if the detected object is a person with high confidence
            if conf > 60 and class_detect in ['Fall Detected', 'NoFall Detected']: 
                cvzone.cornerRect(frame, [x1, y1, x2-x1, y2-y1], l=30, rt=6)
                cvzone.putTextRect(frame, f'{class_detect}', [x1 + 8, y1 - 12], thickness=2, scale=2)
                       
    # Display the frame
    cv2.imshow('frame', frame)
    
    # Break the loop if 't' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('t'):
        break

# Release the webcam capture and destroy all windows
cap.release()
cv2.destroyAllWindows()
```
