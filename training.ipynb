```python
def load_images_and_labels(image_dir, label_dir):
    images = []
    labels = []
    for folder in os.listdir(image_dir):
        for file in os.listdir(os.path.join(image_dir, folder)):
            img_path = os.path.join(image_dir, folder, file)
            label_path = os.path.join(label_dir, folder, file.replace('.jpg', '.txt'))
            
            img = cv2.imread(img_path)
            if img is not None:
                images.append(img)
                
                with open(label_path, 'r') as f:
                    # Read the class label
                    label_line = f.readline().strip()
                    label = int(label_line.split()[0])  # Extract the first element as the label

                    bbox = list(map(float,label_line.split()[1:]))  # Convert the rest of the elements to floats
                    print(bbox)
                    # labels.append([label, bbox])  # Append both label and bounding box to the list
                    
    return np.array(images), np.array(labels)

# # Example usage
image_dir = 'data/images'   
label_dir = 'data/labels'

images, labels = load_images_and_labels(image_dir, label_dir)
```
